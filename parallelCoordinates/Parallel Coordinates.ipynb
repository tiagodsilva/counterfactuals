```python
def sort_parallel(indexes, column, df):
    if len(indexes)==1 or column==df.shape[1]:
        return indexes
    
    feat = df.columns[column]
    
    return_indexes = [] 
    a = df.loc[indexes, feat]
    for value in a.unique(): 
        values = a[a==value]
        return_indexes+=sort_parallel(list(values.index), column+1, df)
    return return_indexes 
```

```python
new_df = df.copy()
step = 0.005
for column in range(new_df.shape[1]):
    feat = new_df.columns[column]
    var = new_df[feat].max()-new_df[feat].min()

    for value in sorted(new_df[feat].unique()):
        values = new_df[feat][new_df[feat]==value]
        if column==0:
            indexes = sort_parallel(list(values.index), column+1, new_df)
        else:
            indexes = sort_parallel(list(values.index), column-1, new_df) 
        steps = step*var*np.arange(values.shape[0])-step*var*values.shape[0]/2
        new_df.loc[indexes,feat] = new_df.loc[indexes,feat]+steps
fig = go.Figure(data=
    go.Parcoords(
        line_color='blue',
        dimensions = [dict(label = label, values = new_df[label])
                      for label in new_df]
    )
)
fig.show() 
new_df 
```
